Making a dataframe of all rooms:

```python
Listroomsnpage = []

for i in range(1,13):
    CC = "#__layout > div > div.listing-layout > main > div > div.plp-view__container > div > div > div > div:nth-child(3) > div.listing-card > div > div:nth-child(" + str(i) +")"
    result1 = driver.find_elements_by_css_selector(CC)
    link = result1[0].find_element_by_css_selector("a").get_attribute('href') 
    name = result1[0].find_element_by_css_selector("p.vertical-card__name").text
    price = result1[0].find_element_by_css_selector("div.pricing").text
    Listroomsnpage.append([name,price,link])
```

```python
df = DataFrame(Listroomsnpage,columns=['Room Name','Price',"Link"])
df
```

```python
list = []
for room in range(len(df.Link)):
    driver.get(df.Link[room])
    star = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp-head.accommodation-pdp__head > div.title > div > div.all-center > span.score')[0].text
    review = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp-head.accommodation-pdp__head > div.title > div > div.all-center > span.count')[0].text
    rule1 = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp__content > div.accommodation-pdp__main-content > div.accommodation-policies > div.accommodation-policies__policies > div.rules > div > div:nth-child(2)')[0].text
    rule2 = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp__content > div.accommodation-pdp__main-content > div.accommodation-policies > div.accommodation-policies__policies > div.rules > div > div:nth-child(3)')[0].text
    rule3 = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp__content > div.accommodation-pdp__main-content > div.accommodation-policies > div.accommodation-policies__policies > div.rules > div > div:nth-child(4)')[0].text
    rule4 = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp__content > div.accommodation-pdp__main-content > div.accommodation-policies > div.accommodation-policies__policies > div.rules > div > div:nth-child(5)')[0].text
    rule5 = driver.find_elements_by_css_selector('#__layout > div > div.pdp > div > div.accommodation-pdp > div.accommodation-pdp__content > div.accommodation-pdp__main-content > div.accommodation-policies > div.accommodation-policies__policies > div.rules > div > div:nth-child(6)')[0].text
    list.append([star, review, rule1, rule2, rule3, rule4, rule5])
```

```python
df1 = DataFrame(list, columns=["Star", "Reviews", "Rule 1", 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5'])
df2 = pd.concat([df,df1],axis=1)
df2
```

```python
df2.to_csv('Rooms.csv')
```
